## 3. Inspect Parsed Page JSON

````python
def choose_product_page(files):
    for path in files:
        if path.name == "bacardi_crawl_summary.json":
            continue
        data = json.loads(path.read_text(encoding="utf-8"))
        source_url = data.get("source_url", "")
        if "/our-rums/" in source_url and source_url.rstrip("/") != "https://www.bacardi.com/our-rums":
            return path
    return None

selected_page_path = choose_product_page(parsed_files)
print("Selected parsed product page:", selected_page_path)
````

````python
if selected_page_path is None:
    print("No parsed product page found. Run the crawl first or add files to data/parsed_products.")
else:
    page_record = json.loads(selected_page_path.read_text(encoding="utf-8"))
    print("source_url:", page_record.get("source_url"))
    print("title:", page_record.get("title"))
    print("h1:", page_record.get("h1"))
    print("\nClean text excerpt:\n")
    print(page_record.get("clean_text", "")[:3000])
    print("\nDiscovered product links:")
    pprint(page_record.get("product_links", []))
````

## 6. Batch ProductCard Parse

This section can parse several product pages with `--limit`. Keep it guarded because it may call the real LLM multiple times.

````python
import warnings

RUN_BATCH_LLM_PARSE = True  # Set True when OPENAI_API_KEY is configured.

if not RUN_BATCH_LLM_PARSE:
    print("Skipping batch LLM parse. Set RUN_BATCH_LLM_PARSE = True to run it.")
elif not os.getenv("OPENAI_API_KEY"):
    print("Skipping batch LLM parse: OPENAI_API_KEY is not set.")
else:
    with warnings.catch_warnings():
        warnings.filterwarnings("ignore", message="Invalid schema.*structured output.*")
        batch_summary = parse_directory(
            input_dir=paths["parsed_products"],
            output_dir=paths["catalog_products"],
            limit=3,
            force=True,
        )

    successful_results = [result for result in batch_summary.results if result.output_file and not result.skipped]
    skipped_results = [result for result in batch_summary.results if result.skipped]

    print(f"Parsed ProductCards: {len(successful_results)}")
    print(f"Skipped files: {len(skipped_results)}")
    print(f"Errors: {len(batch_summary.errors)}")

    if batch_summary.errors:
        print("\nErrors:")
        for error in batch_summary.errors:
            print(" -", error)

    if skipped_results:
        print("\nSkipped:")
        for result in skipped_results:
            print(f" - {Path(result.input_file).name}: {'; '.join(result.warnings)}")

    print("\n=== FINAL PRODUCT CARDS ===")
    for result in successful_results:
        card_path = Path(result.output_file)
        card = json.loads(card_path.read_text(encoding="utf-8"))
        print(f"\n--- {card.get('name', card_path.name)} ---")
        print(json.dumps(card, ensure_ascii=False, indent=2))
````

### BM25 Experiment Over Concise Profile Text

This is a notebook-only experiment. It tests the idea that `searchable_text` may be too verbose for retrieval and that a shorter retrieval text made from display/tasting/tags can rank exact lexical matches more clearly.

````python
import math
from collections import Counter

def bm25_tokenize(text):
    normalized = "".join(ch.lower() if ch.isalnum() else " " for ch in text)
    return [token for token in normalized.split() if len(token) > 1]

def concise_profile_text(profile):
    parts = [
        profile.name,
        profile.category or "",
        profile.display_description or "",
        profile.tasting_summary or "",
        " ".join(profile.flavor_tags),
        " ".join(profile.usage_tags),
        " ".join(profile.cocktail_names),
    ]
    return " ".join(part for part in parts if part).strip()

def bm25_search(query, profiles, top_k=3, k1=1.5, b=0.75):
    docs = [bm25_tokenize(concise_profile_text(profile)) for profile in profiles]
    query_tokens = bm25_tokenize(normalize_query(query))
    avgdl = sum(len(doc) for doc in docs) / max(len(docs), 1)
    doc_freq = Counter()
    for doc in docs:
        for token in set(doc):
            doc_freq[token] += 1

    results = []
    total_docs = len(docs)
    for profile, doc in zip(profiles, docs):
        tf = Counter(doc)
        score = 0.0
        matched = []
        for token in query_tokens:
            if tf[token] == 0:
                continue
            matched.append(token)
            idf = math.log(1 + (total_docs - doc_freq[token] + 0.5) / (doc_freq[token] + 0.5))
            denom = tf[token] + k1 * (1 - b + b * len(doc) / max(avgdl, 1))
            score += idf * (tf[token] * (k1 + 1)) / denom
        results.append((score, profile, sorted(set(matched)), concise_profile_text(profile)))
    return sorted(results, key=lambda item: item[0], reverse=True)[:top_k]

bm25_results = bm25_search(direct_query, profiles, top_k=3)
print("BM25 query:", direct_query)
print("BM25 normalized query:", normalize_query(direct_query))
print("\nBM25 top-k over concise profile text:")
for rank, (score, profile, matched, text) in enumerate(bm25_results, start=1):
    print(f"{rank}. {profile.name} [{profile.product_id}] score={score:.3f}")
    print("   matched tokens:", matched)
    print("   concise_text:", text[:700])
````
